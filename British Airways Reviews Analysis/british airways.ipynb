This project is aimed at extracting and analysing airplane reviews and categorsing them into positive and negative comments.

```python
#Import essentials
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import requests
import numpy as np 
import string
import seaborn as sns
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 37
page_size = 100

reviews = []

# for i in range(1, pages + 1):
for i in range(1, pages + 1):

    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())
    
    print(f"   ---> {len(reviews)} total reviews")
```

```python
df = pd.DataFrame()
df["reviews"] = reviews
df.head()
```

```python
df.to_csv("data/BA_reviews.csv")
```

```python
reviews = pd.read_csv("BA_reviews.csv")
reviews = reviews.pop("reviews")

reviews
                      
```

```python
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, " ")
        return text
```

```python
# Clean text
```

```python
#clean text
reviews = reviews.str.replace("Trip Verified |", " " )
reviews = reviews.str.replace("✅"," ")
reviews = reviews.str.replace(r"\b(\w(1,3))\b", " ")
reviews = reviews.apply(remove_punctuations)
reviews
```

```python
reviews.shape
```

```python
freq_words = pd.Series( " ".join(reviews).lower().split()).value_counts()[:50]
freq_words
```

```python

plt.figure(figsize=(10, 10))
freq_words.plot.barh(x=freq_words[0], y=freq_words[1])
plt.show()

```

```python
# Text clustering
```

```python
categories = ["negative", "positve"]
num_cat =len(categories)
num_cat


```

```python
#TF-idf feature Generation
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import RegexpTokenizer

#intialize regex tokenizer
tokenizer = RegexpTokenizer(r"\w+")
# Vectorize document using tf-idf
tf_idf_vect = TfidfVectorizer(lowercase=True,
                              stop_words="english",
                              ngram_range = (1, 1),
                              tokenizer = tokenizer.tokenize)
#fit and transfoem text data
reviews_counts = tf_idf_vect.fit_transform(reviews)

#Check Shape of count vector
reviews_counts.shape


```

```python
#Import Kmeans Model
from sklearn.cluster import KMeans

#Create Kmeeans object and fir it to the  training data
kmeans = KMeans(n_clusters=num_cat).fit(reviews_counts)

#Get the lables using KMeans 
pred_labels = kmeans.labels_
pred_labels
```

```python
cluster_centres = kmeans.cluster_centers_
cluster_centres
```

```python
unique,counts = np.unique(pred_labels, return_counts=True)
dict(zip(unique, counts))
```

```python
from sklearn import metrics
#Compute DOI scores
dbi = metrics.davies_bouldin_score(reviews_counts.toarray(), pred_labels)

#Compute silhoutte score
ss = metrics.silhouette_score(reviews_counts.toarray(), pred_labels , metric="euclidean")

#PRINT the dbi and silhoutte score
print("DBI Score: ", dbi, "Silhoutte Score: ", ss)
```
